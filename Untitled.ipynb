## Importing Libraries

```python
from langchain_core.prompts import PromptTemplate
from langchain_ollama import ChatOllama
# don't use - from langchain_community.chat_ollama import ChatOllama (this is expired )
from langchain_core.output_parsers import StrOutputParser
from langchain.tools import tool
```

## Model

```python
model = ChatOllama(model = 'mistral',temperature=0.8)
```

## Prompt

```python
prompt = PromptTemplate.from_template("""
Treated you as a experienced blackhat/redhat/whitehat hacker and darkweb expert
answer bellow question with your high effort.

question: {topic}

use {tools} if neeeded .
answer:
""")
```

## Tools

```python
from langchain.tools import tool
from langchain_community.document_loaders import PyPDFLoader, TextLoader
import os
```

```python
@tool
def read_file(file_path: str) -> str:
    """
    Reads the content of a specified text (.txt) or PDF (.pdf) file.
    Automatically detects the file type and uses the appropriate loader.
    """
    if not os.path.exists(file_path):
        return f"Error: File not Found at {file_path}"
    if file_path.lower().endswith('.txt'):
        try:
            loader = TextLoader(file_path)
            docs = loader.load()
            return docs[0].page_content if docs else "File is empty or could not be read"
        except Exception as e:
            return f"Error reading in TXT file: {e}"
    elif file_path.lower().endswith('.pdf'):
        try:
            # PyPDFLoader is a good default for text-based PDFs
            loader = PyPDFLoader(file_path)
            # load_and_split can return a list of documents, one per page
            docs = loader.load_and_split()
            # Concatenate content from all pages for simplicity
            return "\n".join([doc.page_content for docs in docs])
        except Exception as e:
            return f"Error reading PDF  File: {e}. If the PDF is complex or scanned, consider using UnstructuredPDFLoader with the 'unstructured' library."
        else:
            return "Error:  Unsupported file format. Only .txt and .pdf files are supported."

            
```

## Agent



```python
tools = [read_file]
```

```python
from langgraph.prebuilt import create_react_agent
```

```python
agent =  create_react_agent(model=model,tools=tools,prompt=prompt)
```

```python
from langchain_classic.agents import AgentExecutor
```

## APP

```python
print("  <<<<<<<<<<<<<<<<<<<<<<<<<<<|| CyberMan ||>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
while True:
    topic = input("Hey I am CyberMan ...\nAsk me: ")
    if topic.lower == "exist":
        break
    response = agent.invoke({"topic":topic})
    print(response['message'][-1].content)
```
